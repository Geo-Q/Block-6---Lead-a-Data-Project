### JO event per day

```python
df_event_in_jo =pd.read_csv("02-Sites-Stations-Calendar/paris-2024-sites-olympiques-et-paralympiques-franciliens.csv",delimiter=";")
df_event_in_jo["nom"] = df_event_in_jo["nom"].apply(lambda x: x.upper())
df_event_in_jo = df_event_in_jo.loc[:11,:]
df_event_in_jo.head()
```

```python
def new_name(site):
    if site == "ROLAND-GARROS":
        return "EVENT_RG"
    elif site == "PARC DES PRINCES":
        return "EVENT_PDP"
    elif site == "STADE DE FRANCE":
        return "EVENT_SDF"
    elif site == "GRAND PALAIS":
        return "EVENT_GP"
    elif site == "ARENA INVALIDES":
        return "EVENT_INV"
    elif site == "STADE DE LA CONCORDE":
        return "EVENT_CCD"
    elif site == "ARENA CHAMP DE MARS":
        return "EVENT_CDM"
    elif site == "STADE TOUR EIFFEL":
        return "EVENT_TEF"
    elif site == "ARENA PARIS SUD":
        return "EVENT_SUD"
    elif site == "ARENA LA CHAPELLE":
        return "EVENT_CHP"
    elif site == "ARENA BERCY":
        return "EVENT_BER"
    elif site == "STADE YVES DU MANOIR":
        return "EVENT_YDM"
    else:
        return 0


def num_month(month):
    if month == "Juillet":
        return "07"
    elif month == "Ao�t":
        return "08"
    else:
        return 0
```

```python
df_event_in_jo2["nom"] = df_event_in_jo2["nom"].apply(lambda x: new_name(x))
df_event_in_jo2 = df_event_in_jo2.set_index('nom')
df_event_in_jo2 = df_event_in_jo2.T
df_event_in_jo2 = df_event_in_jo2.reset_index()
df_event_in_jo2["index"] = df_event_in_jo2["index"].apply(lambda x: x.split())
df_event_in_jo2["index"] = df_event_in_jo2["index"].apply(lambda x: x[-1:-4:-1])
df_event_in_jo2["index"] = df_event_in_jo2["index"].apply(lambda x: x[0] + "-" + num_month(x[1]) + "-" + x[2])
df_event_in_jo2 = df_event_in_jo2.rename(columns={"index" : 'ds'})
df_event_in_jo2.head()
```

```python
juillet = []
for i in range(23):
    juillet.append("2024-07-" + str(["0" + str(i+1) if (i+1) < 10 else str(i+1)][0]))
df_juillet = pd.DataFrame(juillet, columns=['ds'])
for el in df_event_in_jo2.columns[1:]:
    df_juillet[el] = 0
df_juillet.head()
```

```python
aout = []
for i in range(31-11):
    aout.append("2024-08-" + str(i+12))
df_aout = pd.DataFrame(aout, columns=['ds'])
for el in df_event_in_jo2.columns[1:]:
    df_aout[el] = 0
df_aout.head()
```

```python
event_jo = pd.concat([df_juillet, df_event_in_jo2, df_aout], axis=0, ignore_index=True)
for el in event_jo.columns[1:]:
    event_jo[el] = event_jo[el].apply(lambda x: int(x))
event_jo = event_jo.set_index('ds')
event_jo.to_csv('99-Data_Clean/event_jo.csv')
event_jo
```

### Geoloc with station capacity

```python
geo_jo_org = pd.read_csv('99-Data_Clean/geoloc_jo.csv')
geo_jo = geo_jo_org.loc[geo_jo_org['type']=='station']
geo_jo = geo_jo.reset_index().drop(axis=1, columns='index')
liste_station = list(geo_jo['lieu'])
liste_station
```

```python
liste_moy = []
liste_std = []
liste_satur = []
data_org = pd.read_csv("99-Data_Clean/data_concatenaited.csv")
```

```python
for station in liste_station:
    mask = (data_org['LIBELLE_ARRET'] == station)
    data = data_org.loc[mask]
    data['JOUR'] = data['JOUR'].apply(lambda x: pd.to_datetime(x))
    data = data.sort_values(by='JOUR', ascending=True)
    data['NB_VALD'] = data['NB_VALD'].apply(lambda x: 5 if x == 'Moins de 5' else x)
    data['NB_VALD'] = data['NB_VALD'].apply(lambda x: int(x))
    mask_years=(data['JOUR'].dt.year<2020)
    data=data.loc[mask_years]
    data = data[['JOUR','NB_VALD']]
    data = data.groupby(['JOUR']).sum()
    data = data.reset_index()
    
    mean_data = data['NB_VALD'].mean()
    std_data = data['NB_VALD'].std()
    max_data = mean_data + (std_data * (1.5))
    
    liste_moy.append(int(mean_data))
    liste_std.append(int(std_data))
    liste_satur.append(int(max_data))
    
```

```python
print(liste_moy)
print(liste_std)
print(liste_satur)
```

```python
df = pd.DataFrame(list(zip(liste_station, liste_moy, liste_std, liste_satur)), columns=["Station", "Mean", "Std", "Saturation"])
df
```

```python
geo_jo['capacite'] = df['Saturation']
geo_jo
```

```python
file = open("03-Scraping_Past_Event_in_existing_Site/Scrap_pdp/src/past_events_pdp.json")
results = json.load(file)
results[:3]
```

```python
df_pdp = pd.DataFrame(results)
df_pdp.describe(include='all')
```

```python
df_pdp['affluence'][282] = '39.123'
df_pdp['affluence'] = df_pdp['affluence'].apply(lambda x: x if x != 'X' else '0')
df_pdp = df_pdp.dropna().reset_index()
df_pdp = df_pdp.drop(axis=0, columns='index')
df_pdp.head()
```

```python
df_pdp['affluence'] = df_pdp['affluence'].apply(lambda x: "".join(x.split(".")))
df_pdp['affluence'] = df_pdp['affluence'].astype('int')
df_pdp.info()
```

```python
pdp_mean = df_pdp['affluence'].mean()
pdp_std = df_pdp['affluence'].std()
mask = ((df_pdp['affluence'] > (pdp_mean - (2 * pdp_std))) & (df_pdp['affluence'] < (pdp_mean + (2 * pdp_std))))
df_pdp = df_pdp.loc[mask]
df_pdp.info()
```

```python
df_pdp['datestamp'] = df_pdp['date'].apply(lambda x: x.split('.')[1:])
df_pdp['datestamp'] = df_pdp['datestamp'].apply(lambda x: '20' + x[2] + "-" + x[1] + "-" + x[0])
df_pdp['datestamp'] = df_pdp['datestamp'].apply(lambda x: pd.to_datetime(x))
df_pdp.head()
```

```python
df_pdp.to_csv('99-Data_Clean/df_pdp.csv', index=False)
```

##### Stade de France

```python
file = open("03-Scraping_Past_Event_in_existing_Site/Scrap_sdf/src/past_events_sdf.json")
results = json.load(file)
results[:3]
```

```python
df_sdf = pd.DataFrame(results)
df_sdf['event_name'] = df_sdf['event_name'].apply(lambda x: " ".join(x.split()))
df_sdf['date'] = df_sdf['date'].apply(lambda x: " ".join(x.split()))
df_sdf['event_description'] = df_sdf['event_description'].apply(lambda x: " ".join(x.split()))
df_sdf.head()
```

```python
def month_fr(x):
    if x == 'janvier':
        return '01'
    if x == 'février':
        return '02'
    if x == 'mars':
        return '03'
    if x == 'avril':
        return '04'
    if x == 'mai':
        return '05'
    if x == 'juin':
        return '06'
    if x == 'juillet':
        return '07'
    if x == 'août':
        return '08'
    if x == 'septembre':
        return '09'
    if x == 'octobre':
        return '10'
    if x == 'novembre':
        return '11'
    else:
        return '12'
```

```python
df_sdf['datestamp'] = df_sdf['date'].apply(lambda x: x.split()[1:])
df_sdf['datestamp'] = df_sdf['datestamp'].apply(lambda x: x[2] + "-" + month_fr(x[1]) + "-" + x[0])
df_sdf['datestamp'] = df_sdf['datestamp'].apply(lambda x: pd.to_datetime(x))
df_sdf.head()
```

```python
df_sdf.to_csv('99-Data_Clean/df_sdf.csv', index=False)
```
