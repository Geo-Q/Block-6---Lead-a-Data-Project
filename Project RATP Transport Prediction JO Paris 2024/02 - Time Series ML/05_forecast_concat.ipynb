```python
import pandas as pd
import numpy as np
```

```python
geo_data = pd.read_csv("99-Data_Clean/geoloc_jo24.csv")
geo_data.head()
```

```python
liste_station = list(geo_data['lieu'].loc[geo_data['type']=="station"])
liste_station
```

```python
df_forecast_jo = pd.read_csv(f"98-Forecast/forecast_jo_{liste_station[0]}.csv")
for i in range(len(liste_station) - 1):
    df = pd.read_csv(f"98-Forecast/forecast_jo_{liste_station[i+1]}.csv")
    df_forecast_jo = pd.concat([df_forecast_jo, df], axis=0, ignore_index=True)
df_forecast_jo['yhat']=np.exp(df_forecast_jo['yhat'])
df_forecast_jo = df_forecast_jo.merge(geo_data, left_on='station', right_on='lieu')
df_forecast_jo = df_forecast_jo.drop(axis=1, columns=['lieu', 'type', 'capacite'])
df_forecast_jo.head()
```

```python
df_forecast_jo.to_csv('99-Data_Clean/df_forecast_jo.csv', index=False)
```
